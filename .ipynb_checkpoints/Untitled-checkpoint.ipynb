# exploratory notebook
This notebook is used for initial exploration on grabbing financial data using pandas_datareader.

## imports

```python
import pandas_datareader.data as web
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## grabbing data

```python
tickers = ["AAPL", "MSFT", "IBM"]

start_date = "2014-01-01"
end_date = "2017-12-31"

data = web.DataReader(tickers, data_source="iex", start=start_date, end=end_date)
```

Below we'll sort the data on a per-ticker basis.

```python
aapl = data["AAPL"]
msft = data["MSFT"]
ibm = data["IBM"]
```

```python
aapl.head()
```

## digging into AAPL
Below we'll explore the AAPL pricing data.

```python
type(aapl.index[0])
```

We'll need to convert this date index to a datetime.

```python
aapl.index = pd.to_datetime(aapl.index)
type(aapl.index[0])
```

Let's look at a slice of the pricing data history. For example, just the dates corresponding to 2015, and only the high/low

```python
figsize = (10, 6)
```

```python
aapl2015 = aapl["2015"]
aapl2015["high"].plot(figsize=figsize)
aapl2015["low"].plot(figsize=figsize)
plt.legend()
```

Additionally, we can add a grid to the plot to better visualize certain time breaks.

## dealing with date discrepencies
Suppose we had financial data for 2 companies with different date periods.

```python
aapl_2014_2015 = web.DataReader("AAPL", data_source="iex", start="2014-01-01", end="2015-12-31")
msft_2015_2016 = web.DataReader("MSFT", data_source="iex", start="2015-01-01", end="2016-12-31")
```

For whatever reason, suppose we wanted to figure out all the dates we have pricing data for AAPL that we **don't** have pricing data for MSFT. We could accomplish this by converting the indices to sets, and then taking the set differential (or you could plot it).

### convert indices

```python
aapl_2014_2015.index = pd.to_datetime(aapl_2014_2015.index)
msft_2015_2016.index = pd.to_datetime(msft_2015_2016.index)
```

### plot

```python
aapl_2014_2015["close"].plot()
msft_2015_2016["close"].plot()
```

### set differential

```python
set_aapl_2014_2015_dates = set(aapl_2014_2015.index)
set_msft_2015_2016_dates = set(msft_2015_2016.index)
print(len(list(set_msft_2015_2016_dates - set_aapl_2014_2015_dates)))
```

### finding common ground
However, there clearly is data available for both datasets under a specified time period, as seen in the plot above. In order to get data where pricing data for both is available, we can use join.

```python
aapl_msft_2015 = aapl_2014_2015[["close"]].join(msft_2015_2016[["close"]], how="inner", lsuffix="_aapl", rsuffix="_msft")
aapl_msft_2015.head()
```

```python
aapl_msft_2015.plot()
```
